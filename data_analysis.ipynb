```python
import pandas as pd
```

```python
market_caps = pd.read_csv("data/new/market_caps.csv").set_index("date")
prices = pd.read_csv("data/new/prices.csv").set_index("date")
volumes = pd.read_csv("data/new/volumes.csv").set_index("date")
```

```python
#given dataframe, returns new dataframe with rankings of values in every row
def rank_df(df):
    columns = list(df.columns.values)
    mapper = {}
    for c in columns:
        mapper[c] = c+"_rank"
    
    return df.rank(axis=1, ascending=False).rename(columns=mapper)
```

```python
rank_df(prices)
```
